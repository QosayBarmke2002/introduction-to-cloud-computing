```python
# -*- coding: utf-8 -*-
# ×˜×•×¤×¡ ×¡×˜×•×“× ×˜×™×: ×‘×—×™×¨×”, ×¦×¤×™×™×”, ×•×¢×“×›×•×Ÿ "×ª×•×›× ×™×ª ××”×•×‘×”" ×‘×§×•×‘×¥ JSON

import json, os
from IPython.display import display, HTML, clear_output
import ipywidgets as w

# ×× ×¢×•×‘×“×™× ×¢× Google Drive, ×‘×˜×œ/×™ ××ª ×”×”×¢×¨×•×ª ×‘×©×ª×™ ×”×©×•×¨×•×ª ×”×‘××•×ª:
# from google.colab import drive
# drive.mount('/content/drive')

# × ×ª×™×‘ ×”×§×•×‘×¥ (×©× ×”/×™ ×œ×¤×™ ×”×¦×•×¨×š, ×œ××©×œ ×œ-"/content/drive/My Drive/Colab Notebooks/students.json")
JSON_PATH = "students.json"

# ×™×¦×™×¨×ª ×§×•×‘×¥ ×“×•×’××” ×× ×œ× ×§×™×™× (×¨×§ ×œ×©× ×”×“×’××”)
if not os.path.exists(JSON_PATH):
    with open(JSON_PATH, "w", encoding="utf-8") as f:
        json.dump([
            {
                "FirstName": "Qosay",
                "LastName": "Barmke",
                "Email": "Qosay.Barmke@e.braude.ac.il",
                "Courses": ["cloud", "web"],
                "url": "https://leetcode.com/",
                "FavoriteProgram": ""
            },
            {
                "FirstName": "Ibrahem",
                "LastName": "Amar",
                "Email": "Ibrahem.Amar@e.braude.ac.il",
                "Courses": ["cloud", "web"],
                "url": "https://leetcode.com/",
                "FavoriteProgram": ""
            }
        ], f, ensure_ascii=False, indent=2)

# ×˜×¢×™× ×”
with open(JSON_PATH, "r", encoding="utf-8") as f:
    students = json.load(f)

# ×•×“× ×©×§×™×™× ×©×“×” FavoriteProgram
for s in students:
    s.setdefault("FavoriteProgram", "")

# ×¢×™×¦×•×‘ ××™× ×™××œ×™ ×•× ×¢×™×
display(HTML("""
<style>
.form {border:1px solid #e5e7eb; border-radius:14px; padding:14px; max-width:900px;}
.form h3 {margin:0 0 10px 0; font-size:20px}
.row {display:flex; gap:10px; flex-wrap:wrap; margin-bottom:6px}
.row > div {flex:1 1 260px}
.status-ok {color:#065f46; background:#ecfdf5; border:1px solid #a7f3d0; padding:6px 10px; border-radius:10px;}
.status-err{color:#7f1d1d; background:#fef2f2; border:1px solid #fecaca; padding:6px 10px; border-radius:10px;}
</style>
<div class="form"><h3>ğŸ“‹ ×˜×•×¤×¡ ×¡×˜×•×“× ×˜×™×</h3></div>
"""))

# ×•×™×“×’'×˜×™×
def option_label(s): return f"{s.get('FirstName','')} {s.get('LastName','')} â€” {s.get('Email','')}"

dd = w.Dropdown(options=[(option_label(s), i) for i,s in enumerate(students)],
                description='×¡×˜×•×“× ×˜:', style={'description_width':'90px'}, layout=w.Layout(width='100%'))

first_name = w.Text(description='×©× ×¤×¨×˜×™:', disabled=True, layout=w.Layout(width='100%'))
last_name  = w.Text(description='×©× ××©×¤×—×”:', disabled=True, layout=w.Layout(width='100%'))
email      = w.Text(description='××™××™×™×œ:',   disabled=True, layout=w.Layout(width='100%'))
courses    = w.Text(description='×§×•×¨×¡×™×:',   disabled=True, layout=w.Layout(width='100%'))
url        = w.Text(description='URL:',      disabled=True, layout=w.Layout(width='100%'))
favorite   = w.Text(description='×ª×•×›× ×™×ª ××”×•×‘×”:', placeholder='×›×ª×•×‘/×›×ª×‘×™ ×›××Ÿ...', layout=w.Layout(width='100%'))

btn_update = w.Button(description='×¢×“×›×•×Ÿ', layout=w.Layout(width='140px'))
status     = w.HTML("")

# ××™×œ×•×™ ×©×“×•×ª ×œ×¤×™ ×‘×—×™×¨×”
def fill_fields(idx: int):
    s = students[idx]
    first_name.value = s.get("FirstName","")
    last_name.value  = s.get("LastName","")
    email.value      = s.get("Email","")
    courses.value    = ", ".join(s.get("Courses", []))
    url.value        = s.get("url","")
    favorite.value   = s.get("FavoriteProgram","")

def on_change_student(change):
    if change['name'] == 'value' and change['new'] is not None:
        fill_fields(change['new'])
        status.value = ""

dd.observe(on_change_student, names='value')

# ×©××™×¨×” ×œ×§×•×‘×¥ ×‘×œ×—×™×¦×”
def on_click_update(_):
    try:
        idx = dd.value
        students[idx]["FavoriteProgram"] = favorite.value.strip()
        with open(JSON_PATH, "w", encoding="utf-8") as f:
            json.dump(students, f, ensure_ascii=False, indent=2)
        status.value = f"<div class='status-ok'>×¢×•×“×›×Ÿ ×•× ×©××¨ ×œÖ¾JSON ×¢×‘×•×¨ <b>{students[idx]['FirstName']} {students[idx]['LastName']}</b>.</div>"
    except Exception as e:
        status.value = f"<div class='status-err'>×©×’×™××” ×‘×©××™×¨×”: {e}</div>"

btn_update.on_click(on_click_update)

# ×¤×¨×™×¡×”
row1 = w.HBox([first_name, last_name])
row2 = w.HBox([email, url])
row3 = w.HBox([courses])
row4 = w.HBox([favorite, btn_update])

ui = w.VBox([dd, row1, row2, row3, row4, status])
display(ui)

# ××™×œ×•×™ ×¨××©×•× ×™
fill_fields(dd.value)

```
